### Bienvenido al curso *Intermediate Machine Learning* de Kaggle

Si ya tienes algo de experiencia en **machine learning** y te gustaría aprender cómo mejorar rápidamente la calidad de tus modelos, estás en el lugar adecuado. En este curso acelerarás tu conocimiento en **machine learning** aprendiendo a:

- abordar tipos de datos habituales en conjuntos de datos del mundo real (*missing values*, *categorical variables*),
- diseñar **pipelines** para mejorar la calidad de tu código de machine learning,
- utilizar técnicas avanzadas de **model validation** (*cross-validation*),
- construir modelos **state-of-the-art** ampliamente utilizados para ganar competiciones de Kaggle (*XGBoost*),
- evitar errores comunes e importantes en **data science** (*leakage*).

A lo largo del curso, aplicarás lo aprendido completando ejercicios prácticos con datos reales para cada nuevo tema. Estos ejercicios utilizan datos de la competición **Housing Prices Competition for Kaggle Learn Users**, en la que usarás 79 variables explicativas diferentes (como el tipo de tejado, el número de dormitorios o el número de baños) para predecir el precio de las viviendas.

Medirás tu progreso enviando predicciones a esta competición y observando cómo subes posiciones en el **leaderboard**.
