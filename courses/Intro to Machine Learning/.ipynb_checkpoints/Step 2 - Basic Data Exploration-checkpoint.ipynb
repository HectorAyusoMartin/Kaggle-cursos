# Interpretando el Data Description:


Cuando se muestra la **data description** de un dataset, aparecen 8 valores estadísticos para cada columna. Estos valores ayudan a entender la distribución y calidad de los datos.

## count
Indica cuántas filas tienen valores **no nulos** (non-missing values) en esa columna.  
Los **missing values** pueden aparecer por muchas razones. Por ejemplo, el tamaño del segundo dormitorio no se recoge si la vivienda solo tiene un dormitorio. El tratamiento de missing data se abordará más adelante.

## mean
La **mean** es la media aritmética, es decir, el promedio de los valores de la columna.

## std
La **std** (standard deviation) mide cuánto se dispersan los valores respecto a la media. Una std alta indica valores muy variados; una std baja indica valores más concentrados.

## min, 25%, 50%, 75% y max
Para entender estos valores, imagina que ordenas todos los datos de una columna de menor a mayor:

- **min**: el valor más pequeño.
- **25% (25th percentile)**: el valor por debajo del cual se encuentra el 25% de los datos.
- **50% (50th percentile o median)**: el valor central; la mitad de los datos están por debajo y la otra mitad por encima.
- **75% (75th percentile)**: el valor por debajo del cual se encuentra el 75% de los datos.
- **max**: el valor más grande.

## Resumen
Esta información permite detectar:
- Valores extremos (outliers)
- Presencia de missing values
- La distribución general de cada variable

Es una herramienta clave para entender el dataset antes de entrenar cualquier modelo de machine learning.
