## Introducción
El aprendizaje automático (machine learning) se basa en identificar patrones a partir de datos históricos para realizar predicciones sobre nuevos casos. Aunque puede parecer un concepto básico para quienes ya tienen experiencia en estadística o modelado, constituye la base para construir modelos predictivos más avanzados.

## Escenario del curso
El curso plantea un ejemplo práctico: predecir el valor de viviendas para invertir en el mercado inmobiliario. Mientras que una persona puede basarse en la intuición y la experiencia previa, el aprendizaje automático replica este proceso utilizando datos para extraer patrones de forma sistemática y objetiva.

## Árboles de decisión
El primer modelo que se introduce es el **árbol de decisión**, una técnica sencilla y fácil de interpretar. En su forma más simple, el árbol divide las viviendas en dos categorías y asigna a cada una un precio promedio basado en datos históricos.  
El proceso de aprender estas reglas a partir de los datos se denomina **entrenamiento** o **ajuste del modelo**, y los datos utilizados se llaman **datos de entrenamiento**.

Una vez entrenado, el modelo puede aplicarse a nuevas viviendas para predecir su precio.

## Mejora del modelo
Un árbol de decisión más realista es aquel que refleja mejor la relación entre las características de una vivienda y su precio, por ejemplo, el número de dormitorios. Sin embargo, un modelo demasiado simple no captura otros factores importantes como baños, tamaño del terreno o ubicación.

Para mejorar la precisión, se utilizan árboles **más profundos**, con más divisiones o *splits*, que consideran múltiples variables. Cada recorrido desde la raíz hasta el final del árbol conduce a una **hoja**, que es donde se genera la predicción final del precio.

## Conclusión
Los árboles de decisión permiten comprender cómo los datos influyen en las predicciones y constituyen la base de modelos más avanzados. A partir de aquí, el siguiente paso es analizar los datos reales que se utilizarán para entrenar estos modelos.
